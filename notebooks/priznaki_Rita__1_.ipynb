# Признаки
смотрим на признак, если важен то начинаем его причесывать
сначала приводим его к адекватному виду (например url нельзя нормально запихнуть в catboost, слишком много категорий получится, поэтому вырезаем из url домен), заполняем пропуски, если плохой то отбрасываем

 - **landing_page**: Страница, на которую перейдет пользователь, если кликнет по рекламе
 - **city**: город юзера
 - **ua_type** -Тип user agent-a
 - **ua_device_type** - Тип устройства
 - **user_fraud_state** - Является ли пользователь фродовым
 -  **user_status** статус пользователя (встречался ли нам он ранее и способ детекции)
 - **creative_size** - Размеры рекламы

```python
from urllib.parse import urlparse

def categorize_creative_size(size):
    '''Обработка рекламных баннеров'''
    if size in ['300x250', '240x400', '336x280', '320x50', '320x100', '300x600', '300x300', '250x250', '300x50']:
        return 'Standard'
    elif size in ['728x90', '970x250', '970x90']:
        return 'Banner'
    elif size in ['480x320', '320x480']:
        return 'Mobile'
    elif size == '160x600':
        return 'Skyscraper'
    elif size == '580x400':
        return 'Large'
    else:
        return 'Other'

def categorize_screen_size(size_str, common_sizes):
    '''Функция для категоризации размера экрана'''
    if size_str is None or size_str not in common_sizes:
        return 'Others'
    width, height = map(int, size_str.split('x'))
    # Рассчитываем диагональ в пикселях
    diagonal = (width**2 + height**2) ** 0.5
    
    # Определение категорий в соответствии с диагональю
    if diagonal < 1500:
        return 'Small'
    elif 1500 <= diagonal < 2000:
        return 'Medium'
    elif 2000 <= diagonal < 2500:
        return 'Large'
    else:
        return 'Extra Large'
    
def categorize_viewability(viewability):
    ''' Функция для категоризации видимость рекламного места на сайте'''
    if viewability is None:
        return 'Unknown'
    elif viewability < 50.0:
        return 'Low'
    elif 50.0 <= viewability < 75.0:
        return 'Medium'
    else:
        return 'High'
    
```

```python
%%time
def preprocess_data(df):
    '''Обработка основных признаков'''
    #  Страница, на которую переходит пользователь после клика
    df['landing_page_domain'] = df['landing_page'].apply(lambda x: urlparse(x).netloc if pd.notnull(x) else 'unknown')
    
    # Обработка city: если город встречается более 100 раз  и он миллионик - 1, иначе 0
    million_cities = [
        "Moskva", "Sankt-Peterburg", "Novosibirsk", "Yekaterinburg",
        "Nizhniy Novgorod", "Kazan", "Chelyabinsk", "Omsk", "Samara",
        "Rostov-on-Don", "Ufa", "Krasnoyarsk", "Perm", "Voronezh", "Volgograd"
    ]    # Города миллионики 

    city_counts = df['city'].value_counts()
    df['city_count'] = df['city'].map(city_counts)
    df['category_city'] = df.apply(
        lambda x: 'freq_million_city' if (x['city'] in million_cities and x['city_count'] >= 100) else 'other',
        axis=1
    )
    
    df['advertiser_id'] = df['advertiser_id'].fillna(0).astype(int)
    
    # Статус пользователя может дать понимание о том, новый это пользователь или возвращающийся
    df['user_status'] = df['user_status']
    df['user_detection_type'] = df['user_detection_type']
    
    # Размер и тип рекламы
    df['category_creative_size'] = df['creative_size'].apply(categorize_creative_size)
    df['creative_type'] = df['creative_type']
    
    # Информация об устройстве, с которого производится просмотр рекламы
    size_counts = df['mobile_screen_size'].value_counts()
    frequency_threshold = 10 
    common_sizes = size_counts[size_counts > frequency_threshold].index.tolist()
    df['screen_size_category'] = df['mobile_screen_size'].apply(lambda x: categorize_screen_size(x, common_sizes))
    
    df['ua_device_type'] = df['ua_device_type'].fillna('unknown').astype(str)
    
    #  Историческая видимость рекламного места на сайте 
    df['viewability_category'] = df['historical_viewability'].apply(categorize_viewability)
    return df

df2 = preprocess_data(df)
```
